```python
import pandas as pd
```

```python
rdf = pd.read_csv("../properties_better_list_improved.csv")
rdf
```

```python
rdf = rdf[["0", "1"]]
rdf
```

```python
rdf["1"] = rdf["1"].apply(lambda x: x[1:-1].split(", "))
rdf
```

```python
data = []

for row in rdf.index:
    for name in rdf.loc[row, "1"]:
        data.append(["<http://www.wikidata.org/entity/{}>".format(rdf.loc[row, "0"]), name[1:-1]])
data
```

```python
pd.DataFrame(data, columns=["uri", "label"]).to_csv("../relations_labels.csv", index=None)
```
